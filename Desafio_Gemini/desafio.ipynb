#2. Manipulação e Transformação


Conserte as datas: Note que há formatos diferentes (YYYY-MM-DD, DD/MM/YYYY). Converta tudo para o tipo datetime.

```python
df['data']
```

```python
df['data'] = pd.to_datetime(df['data'], dayfirst=True, format='mixed')
```

```python
print(df['data'].dtype)
```

Crie a coluna "Total": Multiplique preco por quantidade.

```python
df
```

```python
df['total'] = df['preco'] * df['quantidade']
```

```python
df.info()
```

Limpe o Desconto: A coluna desconto é uma string com "%". Transforme-a em um número decimal (ex: 5% vira 0.05) e trate os nulls como 0.

```python
df['desconto'] = df['desconto'].fillna('0%')
```

```python
df['desconto'] = df['desconto'].apply(lambda x: x.replace('%','').strip())
```
